```python
#Trading bot
lot = 0.1
ticker='GBPUSD'
buy_order_type=mt5.ORDER_TYPE_BUY
sell_order_type=mt5.ORDER_TYPE_SELL
point = mt5.symbol_info(symbol).point
buy_price = mt5.symbol_info_tick(symbol).ask 
sell_price = mt5.symbol_info_tick(symbol).bid
sl=price + 100 * point,
tp=price - 100 * point,
sl_pct=0.05
tp_pct=0.1
buy_sl=buy_price * (1-sl_pct)
buy_tp=buy_price * (1+tp_pct)
sell_sl=sell_price * (1+sl_pct)
sell_tp=sell_price * (1-tp_pct)

def create_order(ticker,qty,order_type,price,sl,tp):
    request = {
        "action": mt5.TRADE_ACTION_DEAL,
        "symbol": ticker,
        "volume": qty,
        "type": order_type,
        "price": price,
        "sl": sl,
        "tp": tp,
        "deviation": deviation,
        "magic": 234000,
        "comment": "python open position",
        "type_time": mt5.ORDER_TIME_GTC,
        "type_filling": mt5.ORDER_FILLING_IOC,
    }
    order = mt5.order_send(request)
    return order
```

```python
# longcondition=close>high[1]
# shortcondition=close<low[1]
# closelongcondition=close<close[1]
# closeshortcondition=close>close[1]

for i in range(100):
    ohlc=pd.DataFrame(mt.copy_rates_range('GBPUSD',mt.TIMEFRAME_M1,datetime(2025,3,21),datetime.now()))
    ohlc['time']=pd.to_datetime(ohlc['time'],unit='s')
    print(ohlc)

    current_close = list(ohlc[-1:]['close'])[0]
    last_close = list(ohlc[-2:]['close'])[0]
    last_high = list(ohlc[-2:]['high'])[0]
    last_low = list(ohlc[-2:]['low'])[0]
    longcondition=current_close>last_high
    shortcondition=current_close<last_low
    closelongcondition=current_close<last_close
    closeshortcondition=current_close>last_close

    alreadybuy = False
    alreadysell = False

    try:
        alreadysell = mt5.positions_get()[0]._asdict(['type'])==1
        alreadybuy = mt5.positions_get()[0]._asdict(['type'])==0
    except:
        pass    

    nopositions = len(mt5.positions_get())==0
    #position=mt5.positions_get()[0]._asdict()['ticket']


    if longcondition:
        if nopositions:
            create_order(ticker,lot,buy_order_type,buy_price,buy_sl,buy_tp)
            print('Buy Order Placed')
        if alreadysell:
            close_order(ticker,lot,buy_order_type,buy_price,position)
            print('Sell Position Closed')
            time.sleep(1)
            create_order(ticker,lot,buy_order_type,buy_price,buy_sl,buy_tp)
            print('Buy Order Placed')
        
    if shortcondition:
        if nopositions:
            create_order(ticker,lot,sell_order_type,sell_price,sell_sl,sell_tp)
            print('Sell Order Placed')
            if alreadybuy:
                close_order(ticker,lot,sell_order_type,sell_price,position)
                print('Buy Position Closed')
                time.sleep(1)
                create_order(ticker,lot,sell_order_type,sell_price,sell_sl,sell_tp)
                print('Sell Order Placed')

    try:
        alreadysell = mt5.positions_get()[0]._asdict(['type'])==1
        alreadybuy = mt5.positions_get()[0]._asdict(['type'])==0
    except:
        pass    

    if closelongcondition and alreadybuy:
        position=mt5.positions_get()[0]._asdict()['ticket']
        close_order(ticker,lot,sell_order_type,sell_price,position)
        print('Buy Position Closed')
    if closeshortcondition and alreadysell:
        position=mt5.positions_get()[0]._asdict()['ticket']
        close_order(ticker,lot,buy_order_type,buy_price,position)
        print('Sell Position Closed')    
        
    alreadybuy == False
    alreadysell == False
    time.sleep(60)
```
